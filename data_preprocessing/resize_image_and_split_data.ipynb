```python
# Define input and output directories 
dirname = "paddy"  # Name of the dataset directory
input_dir = f"./{dirname}/"  # Main input directory containing 'train' and 'test' folders 
output_dir = f"./{dirname}_resize/"  # Directory to save augmented images 
file_extension = ".JPG" 
 
# Define augmentation parameters 
resize_dim = (224, 224) 
 
# Function to perform data augmentation on images in a folder 
def augment_images_in_folder(input_folder, output_folder): 
    # Create output folder if it doesn't exist 
    os.makedirs(output_folder, exist_ok=True) 
 
    # Iterate over each image in the input folder 
    for filename in os.listdir(input_folder): 
        if filename.lower().endswith(('.jpg', '.png')):  # Check for valid image files 
            print(f"Augmenting image: {filename}") 
 
            # Load image 
            image_path = os.path.join(input_folder, filename) 
            img = cv2.imread(image_path) 
            resized_img = cv2.resize(img, resize_dim) 
 
            # Split images at the file extension to keep name without it 
            filename_without_ext = os.path.splitext(filename)[0] 
 
            cv2.imwrite(os.path.join(output_folder, f"{filename_without_ext}_resize{file_extension}"), resized_img) 
 
 
# Augment images in both 'train_images' only becasue 'test_images' have no labels, can cannot be used for training. and 
for data_type in ['train_images']: 
    data_folder = os.path.join(input_dir, data_type) 
    if os.path.exists(data_folder): 
        for class_folder in os.listdir(data_folder): 
            input_class_folder = os.path.join(data_folder, class_folder) 
            output_class_folder = os.path.join(output_dir, data_type, class_folder) 
            print(f"\nAugmenting images in folder: {class_folder} ({data_type})") 
            augment_images_in_folder(input_class_folder, output_class_folder) 
 
print("\nData augmentation complete!")
```

# Split dataset into training and validation sets (80% train_val, 20% test)

```python
data_dir = 'C:\\Users\\dnslab_dear\\Desktop\\paddy dataset\\paddy_resize\\train_images'       # Your original folder
output_dir = 'C:\\Users\\dnslab_dear\\Desktop\\paddy dataset\\paddy_resize\\split_data'  # Where you want the train/val folders
split_ratio = 0.8  # 80% for training

# Get all class names (folder names)
classes = [d for d in os.listdir(data_dir) if os.path.isdir(os.path.join(data_dir, d))]

for cls in classes:
    # Path to the current class folder
    cls_path = os.path.join(data_dir, cls)
    
    # Get all images in this class folder
    images = [f for f in os.listdir(cls_path) if os.path.isfile(os.path.join(cls_path, f))]
    
    # Shuffle the images to ensure randomness
    random.shuffle(images)
    
    # Calculate split index
    train_count = int(len(images) * split_ratio)
    
    train_images = images[:train_count]
    val_images = images[train_count:]
    
    # 2. Define and create destination folders
    for folder_type, images_to_copy in [('train_val', train_images), ('test', val_images)]:
        target_dir = os.path.join(output_dir, folder_type, cls)
        os.makedirs(target_dir, exist_ok=True)
        
        # 3. Copy files to disk
        for img in images_to_copy:
            src_path = os.path.join(cls_path, img)
            dst_path = os.path.join(target_dir, img)
            shutil.copy(src_path, dst_path)

print(f"Stratified split complete! Data saved in {output_dir}")
```
